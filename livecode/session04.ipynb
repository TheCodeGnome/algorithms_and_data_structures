```python
def reverse_str(s):
    if len(s) == 0:
        return s
    else:
        return s[-1] + reverse_str(s[:-1])
```

```python
# Using recursion, positive exponent, base > 0; return answer
def power(base, exponent):
    if exponent <= 0 or base <=0:
        return "Base and exponent must be positive numbers."
    if exponent == 1:
        return base

    return base * power(base, exponent - 1)

print(power(2,1))
print(power(2,5))
print(power(5,2))
print(power(5,-1))
print(power(5,0))

```

```python
# use recursion to determine if a value is inside a list
def is_value_inside_list(list, value):
    if len(list) == 0:
        return False
    elif list[0] == value:
        return True
    else:
        return is_value_inside_list(list[1:], value)
    
print(is_value_inside_list([1,2,3,4], 5))
print(is_value_inside_list([1,2,3], 3))
```

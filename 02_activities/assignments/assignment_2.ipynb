#### Starter Code for Question 2

```python
from typing import List

# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right


def bt_path(root: TreeNode) -> List[List[int]]:
    if root is None:
        return []

    if root.left is None and root.right is None:
        return [[root.val]]

    paths = []

    for child in [root.left, root.right]:
        if child is not None:
            for path in bt_path(child):
                paths.append([root.val] + path)

    return paths

rootNode = TreeNode(1)
leaf1 = TreeNode(2)
leaf2 = TreeNode(2)
leaf3 = TreeNode(3)
leaf4 = TreeNode(5)
leaf5 = TreeNode(6)
leaf6 = TreeNode(7)

rootNode.left=leaf1
rootNode.right=leaf2
leaf1.left=leaf3
leaf1.right=leaf4
leaf2.left=leaf5
leaf2.right=leaf6

print(bt_path(rootNode))
```

> For any given list of integers, scan the list and return the first duplicate value. If there are no duplicates, return -1.

```python
# Your answer here
from typing import List

def first_duplicate(nums: List[int]) -> int:
    if len(nums) <= 1:
        return -1

    num = nums[0]
    remaining_nums = nums[1:]

    if not(num in remaining_nums):
        num = first_duplicate(remaining_nums)
    
    return num


print(first_duplicate([3, 1, 4, 2, 5, 1, 6])) # Output: 1
print(first_duplicate([7, 8, 9, 10])) # Output: -1
print(first_duplicate([4, 5, 6, 4, 6])) # Output: 4
```


-   Copy the solution your partner wrote. 


```python
# Your answer 
from typing import List

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for item in nums:
        if item in seen:
            return item
        else:
            seen.add(item)
    return -1
```


-   Explain why their solution works in your own words.


### Your answer here
> Jose has chosen a clear and easily readable loop to solve the problem. By tracking each number coming from the initial list with the second "seen" list, he can verify if the number has occured before. As a result, he is walking the full, increasing, length of seen for each item in nums until there is a duplicate.


-   Explain the problem’s time and space complexity in your own words.


### Your answer here
> Time complexity: The time complexity will be O(n), where n = len(nums). While we may have to walk seen n times, the add and lookup workout to O(1), thus not impacting the time complexity.
>
> Space complexity: In the event of no duplicates, we would walk the full length of nums. That being the case, we have a complexity of O(n).
